```python
# Create a DataFrame to store cities' information and weather conditions
columns=['city_id' , 'city_name', 'latitude', 'longitude', 'Weather trend next 8 days', 'Precipitation next 8 days (in mm)', 'Average temperature next 8 days (in Celsius)', 'Average temperature felt next 8 days (in Celsius)']
df_cities = pd.DataFrame(columns=columns)
df_cities
```

```python
# Create a loop to extract information from nominatim/openstreetmap api url and put it in dataframe df_cities
for i in range(len(cities)) :
    #print("Work done for city: {}".format(cities[i])) 
    # As got index error, add this print line to check which city create index error >> delete Gorges du Verdon and Ariege from initial list
    
    # url where to find cities' information 
    url = 'https://nominatim.openstreetmap.org/search'
    
    # Creating payload variable to add parameters in the query: 
    payload = {
        "city": cities[i],
        "country": "France",
        "format": "json"
    }
    
    # Getting data / information that includes payload as parameters
    r = requests.get(url, params=payload)
    result = r.json()
    
    # Add json content in the dataframe df_cities
    df_cities.loc[i,'city_id'] = i
    df_cities.loc[i,'city_name'] = cities[i]
    df_cities.loc[i, 'latitude'] = result[0]['lat']
    df_cities.loc[i, 'longitude'] = result[0]['lon']
#------------------------------------------------------------------------------------#
    # Create a loop to extract information from openweathermap api url and put it in dataframe df_cities 
        
    # url where to find weather information
    url_weather= "https://api.openweathermap.org/data/2.5/onecall?"

    # Creating payload variable to add parameters in the query - Following the openweathermap api documentation:
    payload_weather = {
        "city": cities[i],
        "lat": df_cities.loc[i, 'latitude'],
        "lon": df_cities.loc[i, 'longitude'],
        "exclude": "current,minutely,hourly,alerts", # Exclusion of the following parameters proposed by the API: current, minutely, hourly and alerts
        #"cnt": 10, # Free subscription limited forecast to today + next 7 days
        "units": "metric", # Allow to have temperature information in Celsius format
        "lang" : "fr", # Allow to have weather information in French
        "appid": "860b5deeeb2418fc6b28ea615644f0cc", # My API key
    }

    r_weather = requests.get(url_weather,params = payload_weather)
    result_weather = r_weather.json()
    #print(result_weather)

    day = []
    for x in result_weather['daily']:
        day.append(x['temp']['day'])
        average_day_temp = sum(day) / float(len(day))
    
    feels_like = []
    for x in result_weather['daily']:
        feels_like.append(x['feels_like']['day'])
        average_feels_like_temp = sum(feels_like) / float(len(feels_like))
        
    probability_precipitation = []
    for x in result_weather['daily']:
        probability_precipitation.append(x['pop'])
        sum_probability_precipitation = sum(probability_precipitation)
    
    description = []
    for x in result_weather['daily']:        
        description.append(x['weather'][0]['description'])
        most_frequent_description = max(set(description), key = description.count)
    
    df_cities.loc[i,'Weather trend next 8 days']= most_frequent_description
    df_cities.loc[i,'Precipitation next 8 days (in mm)']= sum_probability_precipitation
    df_cities.loc[i,'Average temperature next 8 days (in Celsius)']= average_day_temp
    df_cities.loc[i,'Average temperature felt next 8 days (in Celsius)']= average_feels_like_temp
```
